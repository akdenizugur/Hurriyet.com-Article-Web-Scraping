```python
#Veri İşlemleri
import pandas as pd
import numpy as np
#Veri Kazıma
from bs4 import BeautifulSoup as bs
import requests 
import datetime
from IPython.display import clear_output
```

```python
sections = ['https://www.hurriyet.com.tr/yazarlar/tum-yazarlar/']
```

```python
urls = []
for section in sections:
    new_url = section
    print(new_url)
    html = requests.get(new_url).text
    soup = bs(html, "lxml")
    tags = soup.find("div", class_="autohor-list")
    for a in tags.find_all('a', href=True):
        urls.append((section.split("/")[4],a['href']))
```

```python
urldata = pd.DataFrame(urls)
urldata.columns = ["Kategori","Link"]
urldata
```

```python
urldata = urldata.astype('str')
urldata = urldata.drop_duplicates(subset=None, keep="first", inplace=False)
urldata['Link'] = 'https://www.hurriyet.com.tr' + urldata['Link'].astype(str)
urldata.drop(urldata.loc[urldata['Link']=='https://www.hurriyet.com.tr/yazarlar/elif-ergu/'].index, inplace=True)
urldata.drop(urldata.loc[urldata['Link']=='https://www.hurriyet.com.tr/yazarlar/eren-alphan/'].index, inplace=True)
urldata.drop(urldata.loc[urldata['Link']=='https://www.hurriyet.com.tr/yazarlar/fuat-akdag/'].index, inplace=True)
urldata.drop(urldata.loc[urldata['Link']=='https://www.hurriyet.com.tr/yazarlar/umut-gokbayrak/'].index, inplace=True)
urldata.drop(urldata.loc[urldata['Link']=='https://www.hurriyet.com.tr/yazarlar/y-sinan-tanyildiz/'].index, inplace=True)
urldata.reset_index(drop=True, inplace=True)

urldata
```

```python
def GetData(url):

    html = requests.get(url).text
    soup = bs(html, "lxml")

    body_text = soup.findAll("div", class_="author-content")[0].findAll('p')

    body_text_big = ""
    for i in body_text:
        body_text_big = body_text_big +i.text

    header = soup.find("h1", class_="news-detail-title").text
    timestamp = soup.find("time", class_="author-date").text
    yazar = soup.find("span", class_="author-detail-title").text
    return ((url,yazar,header,body_text_big,timestamp))
```

```python
bigdata = []
k = 0
for i in urldataYazar.Link:
    clear_output(wait=True)
    print(k)
    bigdata.append(GetData(i))
    k = k + 1
```

```python
bigdatax = pd.DataFrame(bigdata)
bigdatax.drop([3,4,5,6],inplace=True)
bigdatax.columns = ["YazıLinki", "YazıYazar", "YazıBaşlığı", "Yazıİçeriği", "YazıTarihi"]
bigdatax.index = range(0,len(bigdatax))
bigdatax = bigdatax.sample(frac=1).reset_index(drop=True)
bigdatax['YazıBaşlığı'] = bigdatax['YazıBaşlığı'].str.replace('?','', regex=True).str.replace(':','', regex=True).str.replace('/','', regex=True).str.replace('"','', regex=True).str.replace('“','', regex=True).str.replace('”', '', regex = True).str.replace('‘','', regex=True).str.replace('’','', regex=True).str.replace('-','', regex=True).str.replace('–','', regex=True)
bigdatax
bigdatax
```
