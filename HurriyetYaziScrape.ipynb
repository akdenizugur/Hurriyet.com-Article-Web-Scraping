```python
#Veri İşlemleri
import pandas as pd
import numpy as np
#Veri Kazıma
from bs4 import BeautifulSoup as bs
import requests 
import datetime
from IPython.display import clear_output
```

```python
sections = ['https://www.hurriyet.com.tr/yazarlar/tum-yazarlar/']
```

```python
urls = []
for section in sections:
    new_url = section
    print(new_url)
    html = requests.get(new_url).text
    soup = bs(html, "lxml")
    tags = soup.find("div", class_="autohor-list")
    for a in tags.find_all('a', href=True):
        urls.append((a['href']))
```

```python
urlsdata = []
for i in range(0, len(urls)):
    urldata = 'https://www.hurriyet.com.tr' + urls[i]
    urlsdata.append(urldata)
```

```python
i = 0
yazarLink = []
for i in range(len(urlsdata)):
    yazarLink.append(urlsdata[i] + "?p=")
```

```python
yazarLink[0:5]
```

```python
for i in range(0,len(yazarLink)):
    try:
        Text = requests.get(yazarLink[i]).text
        soup = bs(Text, 'lxml')
        pag = soup.find("ul", class_="pagination").text.replace(' ','').replace('«','').replace('»','')
    except:
        continue
```

```python
print(yazarLink[i] + " ===== " + str(len(pag)))
```

```python
yazarLinkK = []

for yazarL in yazarLink:
    
    for i in range(1, len(pag)):
        try:
            new_url = yazarL + str(i)
            print(new_url)
            html = requests.get(new_url).text
            soup = bs(html, "lxml")
            tags = soup.find("div", class_="highlighted-box-wrapper")
            for a in tags.find_all('a', href=True):
                yazarLinkK.append((a['href']))
        except:
            continue
```

```python
YazarLinkSon = []
for i in range(0, len(yazarLinkK)):
    YazarLinkSon.append("https://www.hurriyet.com.tr" + yazarLinkK[i])
```

```python
YazarLinkSon[0:5]
```

```python
def GetData(url):

    html = requests.get(url).text
    soup = bs(html, "lxml")

    body_text = soup.findAll("div", class_="author-content")[0].findAll('p')

    body_text_big = ""
    for i in body_text:
        body_text_big = body_text_big +i.text

    header = soup.find("h1", class_="news-detail-title").text
    timestamp = soup.find("time", class_="author-date").text
    yazar = soup.find("span", class_="author-detail-title").text
    return ((url,yazar,header,body_text_big,timestamp))
```

```python
Data = []
k = 0
for i in YazarLinkSon:
    try:
        clear_output(wait=True)
        print(k)
        Data.append(GetData(i))
        k = k + 1
    except:
        continue
```

```python
dataframe = pd.DataFrame(Data)
dataframe.drop([3,4,5,6],inplace=True)
dataframe.columns = ["YazıLinki", "YazıYazar", "YazıBaşlığı", "Yazıİçeriği", "YazıTarihi"]
dataframe.index = range(0,len(dataframe))
dataframe = dataframe.sample(frac=1).reset_index(drop=True)
dataframe['YazıBaşlığı'] = dataframe['YazıBaşlığı'].str.replace('?','', regex=True).str.replace(':','', regex=True).str.replace('/','', regex=True).str.replace('"','', regex=True).str.replace('“','', regex=True).str.replace('”', '', regex = True).str.replace('‘','', regex=True).str.replace('’','', regex=True).str.replace('-','', regex=True).str.replace('–','', regex=True)
dataframe
```

```python
file = r'Yazılar/{}.txt'
for i, row in dataframe.iterrows():
    with open(file.format(row['YazıBaşlığı']), 'w',encoding="utf-8") as f:
        f.write("Yazı Yazarıs: " + str(row['YazıYazar']) + "\n")
        f.write("Yazı Başlığı: " + str(row['YazıBaşlığı']) + "\n")
        f.write("Yazı Linki: " + str(row['YazıLinki']) + "\n")
        f.write("Yazı İçeriği: " + str(row['Yazıİçeriği']) + "\n")
        f.write("Yazı Tarihi: " + str(row['YazıTarihi']) + "\n")
```
